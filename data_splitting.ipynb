```python
import os
from shutil import copy2
import random

# for dirs, sub_dirs, files  in os.walk('data/imagenet_images/'):
#     print(dirs)
#     print(sub_dirs)
#     print(files)
```

```python


def create_train_test_list(target_dir):
    random.seed(42)
    target_images = [os.path.join(target_dir, image_path) for image_path in os.listdir(target_dir)]
    train_split = int(len(target_images) * 0.8)
    train_images_list = random.sample(target_images, train_split)
    test_images_list = list(set(target_images).difference(set(train_images_list)))
    return train_images_list, test_images_list

train_images_list, test_images_list = create_train_test_list('data/model_test_images/abbey')
len(train_images_list), len(test_images_list)
```

```python
def copy_images_to_file(img_path_list, target_dir, train=True):
    if train:
        split_dir = 'train'        
    else:
        split_dir = 'test'

    # Copy trainig images
    for image_path in img_path_list:
        image_file_name = os.path.split(image_path)[-1]
        dest_path = os.path.join(target_dir, split_dir, image_dir, image_file_name)
        print(f"copying :\n {image_path} to \n{dest_path}")  
        copy2(image_path, dest_path)
```

```python
# Create a function to move images

data_dir = "data/model_test_images"
target_dir = 'data/model_test_images_split'

for image_dir in os.listdir(data_dir):
    for split_dir in ["train", "test"]:
        os.makedirs(os.path.join(target_dir, split_dir, image_dir), exist_ok=True)

    # Make training and test lists of target images     
    train_images_list, test_images_list = create_train_test_list(os.path.join(data_dir, image_dir))

    # Copy training images
    copy_images_to_file(img_path_list = train_images_list, target_dir = target_dir, train=True)

    # Copying testing images
    copy_images_to_file(img_path_list = test_images_list, target_dir = target_dir, train=False)
```
